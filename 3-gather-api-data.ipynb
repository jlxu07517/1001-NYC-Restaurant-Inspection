```python
def checkdata(dflst):
    t = 0
    tail = dflst[0]['Unnamed: 0'].tail(1).values[0]
    for df in dflst:
        if t == 0:
            t += 1
            continue
        if df['Unnamed: 0'].head(1).values[0] != tail + 1:
            print(t)
            break
        else:
            tail = df['Unnamed: 0'].tail(1).values[0]
            t += 1

    
   
        
    
    
```

```python
dflst = []
for i in ['0','500','1000','1500','2000','2500','4000','4500',
          '5000','5500','7000','7500','8000','8500','9000','9500','10000','10500','11000','11500','12000']:
    dflst.append(pd.read_csv(dir1+i
                            +'.csv'))
checkdata(dflst) 
```
